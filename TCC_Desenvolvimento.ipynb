```python
# Importando as bibliotecas
import pandas as pd
import sklearn 
import matplotlib.pyplot as plt
```

```python
# Armazenando o arquivo csv do dia 20/09 em uma variável 
dataState_day20_08 = pd.read_csv('dados/covid 20-08.csv')
```

```python
# Alterando os nomes das colunas
mapa = {
    "date" : "Data",
    "state" : "Estado",
    "city" : "Cidade",
    "place_type" : "Tipo de local",
    "confirmed" : "Confirmações",
    "deaths" : "Mortes",
    "is_last" : "É a última atualização?",
    "estimated_population" : "População estimada",
    "city_ibge_code" : "Código IBGE",
    "confirmed_per_100k_inhabitants" : "Confirmados/100k hab",
    "death_rate" : "Mortes/confirmados",
}
dataState_day20_08 = dataState_day20_08.rename(columns = mapa)
```

```python
# Apresentando os dados gerais - dia 20 de setembro
# Fazer aparecer todas as linhas do quadro
pd.set_option('display.max_rows', None)
dataState_day20_08
```

```python
# Casos de óbito e casos confirmados no estado - dia 20 de setembro
dataDeathsConfirmation_20_08 = dataState_day20_08[["Data", "Cidade", "Confirmações", "Mortes"]]
# Apresentando os casos de óbitos e casos confirmados
dataDeathsConfirmation_20_08
```

```python
# Selecionando as 5 Cidades com mais casos confirmados
top8Confirmacoes = dataDeathsConfirmation_20_08[dataDeathsConfirmation_20_08['Confirmações'] >= 13310]
```

```python
top8Confirmacoes = top8Confirmacoes.sort_values('Confirmações', ascending=False, ignore_index=True)
top8Confirmacoes
```

```python
exp = [0.2,0,0,0,0,0,0,0]
plt.pie(top8Confirmacoes["Confirmações"], labels=top8Confirmacoes["Cidade"], autopct="%1.1f%%", explode= exp, 
radius=1.5, shadow=True)
plt.show()

# Grafico de casos confirmados
```
